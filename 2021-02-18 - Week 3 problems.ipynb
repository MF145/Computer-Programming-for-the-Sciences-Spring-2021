$$
\frac{dy}{dt} =  y - \frac{1}{2}{{\bf{e}}^{\frac{t}{2}}}\sin \left( {5t} \right) + 5{{\bf{e}}^{\frac{t}{2}}}\cos \left( {5t} \right)\hspace{0.25in}y\left( 0 \right) = 0
$$

$$
y\left( t \right) = {{\bf{e}}^{\frac{t}{2}}}\sin \left( {5t} \right)
$$

```python
y=0
t=0
dt=0.001

S=Storage()
S+=t,y

for i in range(10000):
    
    dy=(y-.5*exp(t/2)*sin(5*t)+5*exp(t/2)*cos(5*t))*dt
    
    y=y+dy
    t=t+dt
    
    S+=t,y
    
    
t,y=S.arrays()
figure(figsize=(5,3))
plot(t,y)

# the solution

ts=t
ys=exp(t/2)*sin(5*t)

plot(ts,ys)

#ylim(-150,150)


xlabel('time')
ylabel('y')    
```

## Harmonic Oscillator

$$\frac{dx}{dt}= v $$

$$ \frac{dv}{dt}= -k\cdot x/m$$

(suggested values k = 2 and m = 1, initial position x = 3 and speed v = 0)

- determine what's a variable and what's a parameter
    - variables: x,v , [t]
    - parameter: k,m
- set initial values of variables
- set the values of the parameter
- set time, and the time step (dt)

- loop the following:
    - for each variable, calculate the small change: dp = a*p*dt
    - update the variable values: x = x+dx
    - update time: t= t+dt
    - do any bookkeeping

```python
x=3
v=0
t=0
dt=0.1

m=1
k=2

S=Storage()
S+=t,x,v

for i in range(100):
    
    dx=v*dt
    dv=-k*x/m*dt
    
    x=x+dx
    v=v+dv
    t=t+dt
    
    S+=t,x,v
    
    
    
t,x,v=S.arrays()
figure(figsize=(5,3))
plot(t,x)
xlabel('time')
ylabel('x')       

figure(figsize=(5,3))
plot(t,v)
xlabel('time')
ylabel('v')       
```
