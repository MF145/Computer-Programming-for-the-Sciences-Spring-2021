```python
def draw_path(step,maximum_distance):
    for i in range(10000):
        angle=rand()*360
        right(angle)
        forward(step)

        x,y=position()
        r=sqrt(x**2+y**2)

        if r>maximum_distance:
            break

    return i
    
```

```python
reset()
maximum_step=draw_path(step=3,maximum_distance=50)
print(maximum_step)
```

```python
reset()
for i in range(5):
    penup()
    goto(0,0)
    pendown()
    pencolor(choice(['red','blue','green','orange','purple','yellow','pink','cyan']))
    maximum_step=draw_path(step=3,maximum_distance=50)
print(maximum_step)
```
