## Initial Conditions

```python
x=8
t=0
dt=0.1
```

## updating the variables

```python
dx=5*dt
x=x+dx
t=t+dt

print(t,x)
```

## putting it together in a loop

```python
x=8
t=0
dt=0.1

print(t,x)

for i in range(100):
    dx=5*dt
    x=x+dx
    t=t+dt

    print(t,x)

```

```python
x=8
t=0
dt=0.1


saved_x=[]
saved_t=[]

saved_x.append(x)
saved_t.append(t)

for i in range(100):
    dx=5*dt
    x=x+dx
    t=t+dt

    saved_x.append(x)
    saved_t.append(t)

```

```python
plot(saved_t,saved_x)
```

```python
x=8
t=0
dt=0.1


saved_x=[]
saved_t=[]

saved_x.append(x)
saved_t.append(t)

for i in range(100):
    dx=5*t*dt
    x=x+dx
    t=t+dt

    saved_x.append(x)
    saved_t.append(t)

plot(saved_t,saved_x)
```

```python
v=5
x=8
t=0
dt=0.1


saved_v=[]
saved_x=[]
saved_t=[]

saved_v.append(v)
saved_x.append(x)
saved_t.append(t)

for i in range(100):
    dv=5*dt
    dx=v*dt
    x=x+dx
    v=v+dv
    t=t+dt

    saved_v.append(v)
    saved_x.append(x)
    saved_t.append(t)

figure()
plot(saved_t,saved_x)

figure()
plot(saved_t,saved_v)
```
