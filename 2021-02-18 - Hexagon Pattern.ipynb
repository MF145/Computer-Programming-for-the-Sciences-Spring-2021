```python
from sci378.mplturtle import *
```

```python
def hexagon():
    for i in range(6):
        forward(50)
        right(60)
```

```python
reset(figsize=(5,5))
hexagon()

```

```python
reset(figsize=(5,5))

for i in range(60):
    penup()
    forward(400)
    pendown()
    
    hexagon()
    penup()
    backward(400)
    pendown()
    
    left(6)
```

```python
def circle_of_hexagons(size):
    for i in range(60):
        penup()
        forward(size)
        pendown()

        hexagon()
        
        penup()
        backward(size)
        pendown()

        left(6)
```

```python
reset()
circle_of_hexagons(400)

circle_of_hexagons(200)
```

```python
reset()

for size,color in zip([400,300,200,100],["red","green","blue","purple"]):
    print(size,color)
    pencolor(color)
    circle_of_hexagons(size)
```
