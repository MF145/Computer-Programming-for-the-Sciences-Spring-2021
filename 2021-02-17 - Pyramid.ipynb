```python
def square(size):
    for i in range(4):
        forward(size)
        right(90)
```

```python
reset()
square(10)
```

```python
reset()

size=10

for i in range(12):
    square(size)
    forward(size)

pencolor("red")
backward(11*size+size/2)
left(90)
forward(size)
right(90)

for i in range(11):
    square(size)
    forward(size)

pencolor("blue")
backward(10*size+size/2)
left(90)
forward(size)
right(90)

for i in range(10):
    square(size)
    forward(size)
```

```python
reset()

squares_in_a_row=12
size=10

for i in range(12):

    for i in range(squares_in_a_row):
        square(size)
        forward(size)

    penup()
    backward((squares_in_a_row-1)*size+size/2)
    left(90)
    forward(size)
    right(90)
    pendown()
    
    squares_in_a_row=squares_in_a_row-1
```
