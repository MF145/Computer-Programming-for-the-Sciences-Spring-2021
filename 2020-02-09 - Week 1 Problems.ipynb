```python
def factorial(N):
    if N<0:
        raise ValueError("Bad!")
    
    
    value=1
    for i in range(1,N+1):
        value=value*i
        
    return value
```

factorial $N! = 1 \cdot 2 \cdot 3 \cdots N$

```python
factorial(5)
```

```python
factorial(0)
```

```python
factorial('bob')
```

Outline of nested squares:

- draw a square of a given size
- draw a little square
- move over 
- draw a larger square
- move ove
- etc...

```python
from sci378.mplturtle import *
```

```python
def square(size):
    for i in range(4):
        forward(size)
        right(90)
        
def move_over(step):
    penup()
    left(90)
    forward(step//2)
    left(90)
    forward(step//2)
    right(180)
    pendown()    
```

```python
reset()

for i in range(1,8):
    step=8
    size=step*i
    square(size)
    move_over(step)
```

```python
reset()

color=0  # red,  1 for blue
for i in range(5):
    for j in range(5):
        
        if color==0:
            pencolor("red")
            color=1
        else:
            pencolor("blue")
            color=0
            
        
        square(10)
        
        penup()
        forward(15)
        pendown()
        
    penup()
    right(90)
    forward(15)
    right(90)
    forward(15*5)
    pendown()
    right(180)
```
